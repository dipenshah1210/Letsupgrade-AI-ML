```
# 1. Write a program to subtract two complex numbers in python.
a = 3 + 7j
b = 1 + 14j
print("Substract of two complex numbers =",a-b)
```

```
# 2. Write a program to find the fourth root of number.
num = int(input("Enter the number: "))
fourth_rt = num**(1/4)
print(f"Fourth root of {num} is {round(fourth_rt,2)}")
```

```
# 3. Write a program to swap two numbers in python by using temporary variable.
a = input("Enter first number:")
b = input("Enter second number:")
temp=a
a=b
b=temp
print("Value of a:",a)
print("Value of b:",b)
```

```
# 4. Write a program to swap two numbers in python without using temporary variable.
a = int(input("Enter value of first variable:"))
b = int(input("Enter value of second variable:"))
a=a+b
b=a-b
a=a-b
print("a is:",a,"b is:",b)
```

```
 # 5. Write a program to convert fahrenheit to kelvin and celsius both.
 fahrenheit = int(input("Enter the temperature in fahrenheit: "))
 degree_sign= u'\N{DEGREE SIGN}'
 celsius = (fahrenheit - 32)*(5/9)
 kelvin = celsius + 273.15
 print(f"{fahrenheit}{degree_sign} Fahrenheit is {round(kelvin, 2)}{degree_sign} Kelvin"
 f" and {round(celsius, 2)}{degree_sign} Celsius")
```
