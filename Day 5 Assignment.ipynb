**1. Write a python program to find the first 20 non-even prime natural numbers.**

```
ls = []
lsl = []
n = 0
for i in range(1,100,2):
  if n<20:
    for j in range(i, 0, -1):
      if i%j==0:
        ls.append(j)
    if len(ls) == 2:
      lsl.append(i)
      n+=1
  ls.clear()
print("First 20 odd prime numbers:=",lsl)
```

**2. Write a python program to implement 15 functions of string.**

```
ds = 'diPen12'

# Converts first character into upper case
print(f"Output of 'ds.capitalize()' is {ds.capitalize()}")

# Converts string into lower case
print(f"Output of 'ds.casefold()' is {ds.casefold()}")

# Returns the number of times a specified value occurs in string
print(f"Output of 'ds.count('i')' is {ds.count('i')}")

# Returns encoded version of string
print(f"Output of 'ds.encode()' is {ds.encode()}")

# Returns true if the string ends with specified value
print(f"Output of 'ds.endswith('i')' is {ds.endswith('i')}")

# Searches the string for specified value and returns the position
print(f"Output of 'ds.find('e')' is {ds.find('e')}")

# Formats specified value in string
print(f"Output of 'ds.format()' is {ds.format()}")

# Converts string into lower case
print(f"Output of 'ds.lower()' is {ds.lower()}")

# Converts string into upper case
print(f"Output of 'ds.upper()' is {ds.upper()}")

# Returns True if all characters in string are lowercase
print(f"Output of 'ds.islower()' is {ds.islower()}")

# Returns True if all characters in string are digits
print(f"Output of 'ds.isdigit()' is {ds.isdigit()}")

# Returns True if string is an identifier
print(f"Output of 'ds.isidentifier()' is {ds.isidentifier()}")

# Returns True if all characters in string are numeric
print(f"Output of 'ds.isnumeric()' is {ds.isnumeric()}")

# Converts first character of each word into upper case
print(f"Output of 'ds.title()' is {ds.title()}")

# Returns True if all characters in string are alphanumeric
print(f"Output of 'ds.isalnum()' is {ds.isalnum()}")
```

**3.Write a python program to check if the given string is palindrome or Anagram or None of them. Display the message accordingly to the user.**

```
a = input("Enter the first string: ")
b = input("Enter the second string: ")

if a == b[::-1]:
  print(f"{a} and {b} are palindrome")
elif sorted(a) == sorted(b):
  print(f"{a} and {b} are Anagram")
else:
  print(f"{a} and {b} are neither palindrome nor anagram")
```

**4. Write a python's user defined function that removes all the additional characters from the string and converts it finally into lower case using built-in lower()**

```
def stri_con(ds):
  for i in ds:
    if not(ord(i)>=65 and ord(i)<=90 or ord(i)>=97 and ord(i)<=122 or
           ord(i)==64):
      ds = ds.replace(i,"")
    elif ord(i) == 64:
      ds = ds.replace(i, " ")
  return ds.lower()

ds = input("Enter the string: ")
print(stri_con(ds))
```
